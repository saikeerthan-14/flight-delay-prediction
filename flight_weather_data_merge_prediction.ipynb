```python
import pandas as pd
pd.set_option('display.max_columns', None)

types = ["Initial", "Final"]
models = ["m1", "m2"]


model = "m2"
type1 = "Final"
```

```python
ini_df = pd.read_csv(f"./CIS_662_{type1}_{model}.csv")
ini_df
```

```python
ini_df["scheduled_departure_time"] = ini_df["DATE"] + ' ' + ini_df["DEPARTURE TIME"]
ini_df["scheduled_arrival_time"] = ini_df["DATE"] + ' ' + ini_df["ARRIVAL TIME"]
ini_df["Carrier Code"] = ini_df["FLIGHT NUMBER"].apply(lambda x: x.split()[0])
```

```python
ini_df["scheduled_arrival_time"] = pd.to_datetime(ini_df["scheduled_arrival_time"], infer_datetime_format=True)
ini_df["scheduled_departure_time"] = pd.to_datetime(ini_df["scheduled_departure_time"], infer_datetime_format=True)
ini_df
```

```python
if type1 == "Initial":
    syr_df = pd.read_csv("./data/modified/SYR_weather_data_forecast_i.csv")
    jfk_df = pd.read_csv("./data/modified/JFK_weather_data_forecast_i.csv")
    mco_df = pd.read_csv("./data/modified/MCO_weather_data_forecast_i.csv")
    ord_df = pd.read_csv("./data/modified/ORD_weather_data_forecast_i.csv")
elif type1 == "Final":
    syr_df = pd.read_csv("./data/modified/SYR_weather_data_forecast_f.csv")
    jfk_df = pd.read_csv("./data/modified/JFK_weather_data_forecast_f.csv")
    mco_df = pd.read_csv("./data/modified/MCO_weather_data_forecast_f.csv")
    ord_df = pd.read_csv("./data/modified/ORD_weather_data_forecast_f.csv")


weather_factors = ['app_temp', 'clouds', 'dewpt', 'pres', 'rh', 'temp', 'timestamp_local', 'uv', 'vis','weather.code', 'wind_dir', 'wind_gust_spd', 'wind_spd']

```

```python
syr_df["timestamp_local"] == pd.to_datetime(syr_df["timestamp_local"])
jfk_df["timestamp_local"] == pd.to_datetime(jfk_df["timestamp_local"])
mco_df["timestamp_local"] == pd.to_datetime(mco_df["timestamp_local"])
ord_df["timestamp_local"] == pd.to_datetime(ord_df["timestamp_local"])
syr_df = syr_df[weather_factors].sort_values("timestamp_local")
jfk_df = jfk_df[weather_factors].sort_values("timestamp_local")
mco_df = mco_df[weather_factors].sort_values("timestamp_local")
ord_df = ord_df[weather_factors].sort_values("timestamp_local")
```

```python
syr_df["date_time_map"] = pd.to_datetime(syr_df["timestamp_local"])
jfk_df["date_time_map"] = pd.to_datetime(jfk_df["timestamp_local"])
mco_df["date_time_map"] = pd.to_datetime(mco_df["timestamp_local"])
ord_df["date_time_map"] = pd.to_datetime(ord_df["timestamp_local"])
syr_df
jfk_df
mco_df
ord_df
```

```python
print(jfk_df.shape)
print(mco_df.shape)
print(ord_df.shape)
print(syr_df.shape)
```

```python
ini_df["date_time"] = pd.to_datetime(ini_df["scheduled_arrival_time"])
ini_df
```

```python
syr_w_merged_data = pd.merge_asof(ini_df.sort_values('date_time'),
                                       syr_df.sort_values('date_time_map'),
                                       left_on='date_time', right_on='date_time_map', direction='nearest')

syr_w_merged_data.head()
```

```python
syr_w_merged_data
```

```python
jfk_departures_df = syr_w_merged_data[syr_w_merged_data["ORIGIN"]=="JFK"]
jfk_departures_df
```

```python
jfk_w_merged_data = pd.merge_asof(jfk_departures_df.sort_values('date_time'),
                                       jfk_df.sort_values('date_time_map'),
                                       left_on='date_time', right_on='date_time_map', direction='nearest')

jfk_w_merged_data.head()
```

```python
mco_departures_df = syr_w_merged_data[syr_w_merged_data["ORIGIN"]=="MCO"]
mco_departures_df
```

```python
mco_w_merged_data = pd.merge_asof(mco_departures_df.sort_values('date_time'),
                                       mco_df.sort_values('date_time_map'),
                                       left_on='date_time', right_on='date_time_map', direction='nearest')

mco_w_merged_data.head()
```

```python
ord_departures_df = syr_w_merged_data[syr_w_merged_data["ORIGIN"]=="ORD"]
ord_departures_df
```

```python
ord_w_merged_data = pd.merge_asof(ord_departures_df.sort_values('date_time'),
                                       ord_df.sort_values('date_time_map'),
                                       left_on='date_time', right_on='date_time_map', direction='nearest')

ord_w_merged_data.head()
```

```python
final_merged_df = pd.concat([jfk_w_merged_data, mco_w_merged_data, ord_w_merged_data]).sort_values("date_time_map_y").reset_index(drop=True)
final_merged_df
```

```python
rename_cols = {
    "ORIGIN" : "Origin Airport"
}
final_merged_df.rename(columns=rename_cols, inplace=True)
```

```python
final_merged_df["Scheduled Elapsed Time (Minutes)"] = (final_merged_df["scheduled_arrival_time"] - final_merged_df["scheduled_departure_time"]).dt.total_seconds() // 60
final_merged_df
```

```python
final_merged_df.columns
```

```python
drop_cols = ["DATE", "DAY", "FLIGHT NUMBER", 'DEPARTURE TIME',
       'ARRIVAL TIME', 'ARRIVAL STATUS', 'ARRIVAL STATUS_Prev_flight_early',
       'ARRIVAL STATUS_Prev_flight_ontime', 'ARRIVAL STATUS_Prev_flight_late',  'date_time', 'timestamp_local_x', 'timestamp_local_y', 'date_time_map_x',
            'date_time_map_y']
```

```python
final_merged_df.drop(columns=drop_cols, inplace=True)
```

```python
final_merged_df.to_csv(f"./data/modified/{type1.lower()}_predictions_{model}.csv", index=False)
```
