```python
import os

cwd = os.getcwd() 

#departures
cwd += "/data/departures"
departure_df_list = []
for org in os.listdir(cwd):
    if not org.startswith("."):
        path = f"{cwd}/{org}"
        print(path)
        for file in os.listdir(path):
            if not file.startswith("."):
                file_path = f"{path}/{file}"
                temp_df = pd.read_csv(file_path, skiprows=7, skipfooter=1, engine="python")
                temp_df = temp_df[temp_df["Destination Airport"] == "SYR"]
                temp_df["Origin Airport"] = org
                departure_df_list.append(temp_df)
temp_df
```

```python
# temp_df[temp_df["Destination Airport"] == "SYR"]
temp_df
```

```python
departures_df = pd.concat(departure_df_list).reset_index(drop=True)
departures_df
```
