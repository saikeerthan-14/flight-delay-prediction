```python
test_data = pd.read_csv("./data/modified/test.csv")
test_data
```

```python
# Convert date-time columns to datetime type
test_data['scheduled_arrival_time'] = pd.to_datetime(test_data['scheduled_arrival_time'])
test_data['actual_arrival_time'] = pd.to_datetime(test_data['actual_arrival_time'])

# Extract hour and day from the datetime columns
test_data['scheduled_arrival_hour'] = test_data['scheduled_arrival_time'].dt.hour
test_data['actual_arrival_hour'] = test_data['actual_arrival_time'].dt.hour
test_data['day_of_week'] = test_data['scheduled_arrival_time'].dt.day_name()

# Create a new feature for flight status based on arrival delay
def categorize_status(delay):
    if delay <= -5:
        return 'early'
    elif -5 < delay <= 5:
        return 'ontime'
    else:
        return 'late'

test_data['calculated_status'] = test_data['Arrival Delay (Minutes)'].apply(categorize_status)

# Display updated data to verify changes
test_data[['scheduled_arrival_time', 'actual_arrival_time', 'scheduled_arrival_hour', 'actual_arrival_hour', 'day_of_week', 'calculated_status']].head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.preprocessing import LabelEncoder

# Selecting features and target variable
features = test_data[['Carrier Code', 'Origin Airport', 'Scheduled Elapsed Time (Minutes)', 
                      'scheduled_arrival_hour', 'actual_arrival_hour', 'day_of_week']]
target = test_data['calculated_status']

# Encode categorical variables
label_encoder = LabelEncoder()
features['Carrier Code'] = label_encoder.fit_transform(features['Carrier Code'])
features['Origin Airport'] = label_encoder.fit_transform(features['Origin Airport'])
features['day_of_week'] = label_encoder.fit_transform(features['day_of_week'])

# Split the data
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Initialize and train the logistic regression model
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# Predictions and Evaluation
y_pred = model.predict(X_test)
classification_report_output = classification_report(y_test, y_pred)

classification_report_output
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Initialize the StandardScaler
scaler = StandardScaler()

# List of numerical features to scale
numerical_features = ['Scheduled Elapsed Time (Minutes)', 'scheduled_arrival_hour', 'actual_arrival_hour']

# Scale these features
features[numerical_features] = scaler.fit_transform(features[numerical_features])

# Check class distribution
class_distribution = target.value_counts(normalize=True)

# Train a Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42, class_weight='balanced')

# Split the data with scaled features
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Fit the model
rf_model.fit(X_train_scaled, y_train)

# Predictions and Evaluation
y_pred_rf = rf_model.predict(X_test_scaled)
classification_report_rf = classification_report(y_test, y_pred_rf)

class_distribution, classification_report_rf
```
