```python
corr = weather_df.corr(numeric_only=True)
```

```python
corr_pairs = corr.unstack().sort_values(ascending=False).drop_duplicates().head(15)
corr_pairs
```

```python
weather_corr_columns = set()
for pair in corr_pairs.index:
    weather_corr_columns.add(pair[0])
    weather_corr_columns.add(pair[1])
weather_corr_columns
```
